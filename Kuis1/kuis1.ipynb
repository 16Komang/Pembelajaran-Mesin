```python
import pandas as pd
import numpy as np
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
#perintah untuk membaca file csv
df = pd.read_csv('insurance.csv')
df.head()
```

```python
label_sex = {
    'male': 1,
    'female': 0
}

label_smoker ={
    'yes' : 1,
    'no' : 0
}

label_region ={
    'northwest' : 0,
    'southwest' : 1,
    'southeast' : 2,
    'northeast' : 3 
}

#code untuk tabel sex, smoker, region
df['sex'] = df['sex'].map(label_sex)
df['smoker'] = df['smoker'].map(label_smoker)
df['region'] = df['region'].map(label_region)

#untuk menampilkan csv
df.head()
```

```python
#HINTS
from statsmodels.stats.outliers_influence import variance_inflation_factor

#untuk dataset
X = df[['age', 'bmi', 'region','sex', 'smoker', 'children']]
  
#untuk vif data frame
vif_data = pd.DataFrame()
vif_data["feature"] = X.columns
  
#untuk menghitung data vif
vif_data["VIF"] = [variance_inflation_factor(X.values, i)
                          for i in range(len(X.columns))]
  
print(vif_data)
```

```python
#mengeset dataset setelah dimodif
databaru = df
databaru.head()
```

```python
X = databaru.iloc[:, :-1].values
y = databaru.iloc[:, 4].values
```

```python
Xm_train, Xm_test, ym_train, ym_test = train_test_split(X, y, test_size=0.2, random_state=100)

y = y.reshape(len(y), 1)
y.shape
```

```python
mlr = LinearRegression()
mlr.fit(Xm_train, ym_train)

ym_pred = mlr.predict(Xm_test)

gbg = np.concatenate((ym_test, ym_pred))
gbg
```

```python
#untuk menghitung R2 Score
r2_mlr = r2_score(ym_test, ym_pred)
print(r2_mlr)
```
